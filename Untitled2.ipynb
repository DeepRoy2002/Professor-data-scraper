```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options as FirefoxOptions
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup


firefox_options = FirefoxOptions()
firefox_options.headless = True  # Use headless mode
firefox_options.binary_location = r"C:/Program Files/Mozilla Firefox/firefox.exe"


gecko_driver_path = r"C:/Users/anirb/VLSI PROJECT HUNT/geckodriver.exe"
service = Service(executable_path=gecko_driver_path)


driver = webdriver.Firefox(service=service, options=firefox_options)

driver.get("https://iiscprofiles.irins.org/searchc/search/search/")


all_professor_profiles = []


wait = WebDriverWait(driver, 10)

while True:
    try:
        
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "btn-u-dark-blue")))
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, "html.parser")
        professor_profiles = soup.find_all("a", class_="btn-u btn-u-dark-blue pull-right")
        if not professor_profiles:
            print("No more professor profiles found.")
            break

        for profile in professor_profiles:
            profile_link = profile['href']
            profile_text = profile.text.strip()
            all_professor_profiles.append({"link": profile_link, "name": profile_text})

        
        next_page_link = driver.find_element(By.XPATH, "//a[contains(@class, 'btn-u-dark-blue') and contains(text(), 'Next')]")
        next_page_link.click()

    except Exception as e:
        print("An error occurred:", e)
        break
        


for profile in all_professor_profiles:
    print(f"Professor Profile Link: {profile['link']}")
    print(f"Professor Name: {profile['name']}")
    print("-" * 50)

# Close the Selenium driver
driver.quit()
```

```python
import csv

def save_professor_data_to_csv(professor_data, csv_filename):
    """
    Save professor data to a CSV file.

    Args:
        professor_data (list): A list of dictionaries containing professor data.
        csv_filename (str): The name of the CSV file to be created or overwritten.
    """
    try:
        with open(csv_filename, mode="w", newline="", encoding="utf-8") as csv_file:
            fieldnames = [
                "Name of the faculty",
                "Name of the institute",
                "Name of the department",
                "Area of expertise",
                "Institute from the Faculty has done PhD",
                "Contact Email",
                "Contact Number",
                "Contact Mobile Number",
                "H-Index",
                "I-Index"
            ]
            writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

            writer.writeheader()

            for professor in professor_data:
                writer.writerow({
                    "Name of the faculty": professor.get("Name of the faculty", ""),
                    "Name of the institute": professor.get("Name of the institute", ""),
                    "Name of the department": professor.get("Name of the department", ""),
                    "Area of expertise": professor.get("Area of expertise", ""),
                    "Institute from the Faculty has done PhD": professor.get("Institute from the Faculty has done PhD", ""),
                    "Contact Email": professor.get("Contact Email", ""),
                    "Contact Number": professor.get("Contact Number", ""),
                    "Contact Mobile Number": professor.get("Contact Mobile Number", ""),
                    "H-Index": professor.get("H-Index", ""),
                    "I-Index": professor.get("I-Index", "")
                })

        print(f"Professor data saved to {csv_filename}")
    except Exception as e:
        print(f"An error occurred while saving data to {csv_filename}: {e}")

# Example usage:
# save_professor_data_to_csv(all_professor_data, "professor_data.csv")

save_professor_data_to_csv(all_professor_data, 'prof_details.csv')
```
