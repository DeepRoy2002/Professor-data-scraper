```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
# Define the URL of the IISc faculty page
url = "https://irins.org/irins/a/searchc/search"

# Send an HTTP GET request to the URL
response = requests.get(url)


```

```python
# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Create lists to store faculty information
    faculty_names = []
    departments = []
    areas_of_expertise = []
    phd_institute = []
    contact_emails = []
    contact_numbers = []
    contact_mobile_numbers = []

    # Find and extract faculty information
    faculty_list = soup.find_all("div", class_="faculty-details")
    for faculty in faculty_list:
        name = faculty.find("div", class_="faculty-name").text.strip()
        department = faculty.find("div", class_="faculty-department").text.strip()
        expertise = faculty.find("div", class_="faculty-expertise").text.strip()
        phd_inst = faculty.find("div", class_="faculty-phd").text.strip()
        contact_info = faculty.find("div", class_="faculty-contact").text.strip()

        # Extract email, phone, and mobile numbers
        email = contact_info.split("\n")[0].strip()
        phone = contact_info.split("\n")[1].strip()
        mobile = contact_info.split("\n")[2].strip()

        # Append the extracted data to the respective lists
        faculty_names.append(name)
        departments.append(department)
        areas_of_expertise.append(expertise)
        phd_institute.append(phd_inst)
        contact_emails.append(email)
        contact_numbers.append(phone)
        contact_mobile_numbers.append(mobile)

    # Create a DataFrame to store the data
    data = {
        "Name of the faculty": faculty_names,
        "Name of the institute": "Indian Institute of Science (IISc) Bangalore",
        "Name of the department": departments,
        "Area of expertise": areas_of_expertise,
        "Institute from the Faculty has done PhD": phd_institute,
        "Contact Email": contact_emails,
        "Contact Number": contact_numbers,
        "Contact Mobile Number": contact_mobile_numbers,
    }

    faculty_df = pd.DataFrame(data)

    # Export the data to a CSV file
    faculty_df.to_csv("iisc_faculty_data.csv", index=False)

    print("Scraping and data export completed successfully.")
else:
    print("Failed to retrieve the webpage. Check the URL or your internet connection.")
```
